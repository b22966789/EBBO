```python
import math
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
from math import sqrt
```

```python
# 使用 OpenCV 讀取圖檔
img_bgr = cv2.imread('img1.jpg')

# 將 BGR 圖片轉為 RGB 圖片
img_rgb = img_bgr[:,:,::-1]
plt.imshow(img_rgb)
plt.show()
```

```python
back = cv2.imread('back.jpg')
backrgb = back[:,:,::-1]
plt.imshow(backrgb)
plt.show()
```

```python
def change_background(image, new_background, replace_color):
  height, width, channels = image.shape
  for i in range(height):
    for j in range(width):
      if color_distance(image[i][j], replace_color) < 163:
        image[i][j]=new_background[i][j]

  return None
```

```python
def color_distance(color1, color2):
  a = (color1[0]-color2[0])**2
  b = (color1[1]-color2[1])**2
  c = (color1[2]-color2[2])**2
  dist = sqrt(a+b+c)
  return dist
```

```python
replace_color=[255, 255, 255]
change_background(img_rgb, backrgb, replace_color)
```
